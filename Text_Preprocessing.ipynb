# Tokenization

***Tokenization:*** The process of converting given paragraph(corpus) into sentences or words or root word

***Stemming:*** Word representation in stemming doesnot have any useful meaning. It doesnt took much time to process, it just use to find base word or root word.
    
    Example:
    a) history and historical will convert to 'histori'
    b) finally, final, finalized then  we get 'fina' after applying stemming
    
   Application: Stemming is used to analysing sentiment analysis, spam classifier, rating and review application etc
    
***Lemmatization:*** Lemmatization also does the same thing as stemming but it convert into meaningful word.It is easily understand by human being.
    In lemmatization, word representation have meaning to understand. and it took alot of time to process.
    
    Example:
    a) history, historical into history
    b) finally, final, finalized into final
    
   Application: Lemmatization is used in chatbot, question answer application etc


## Load  libraries

```python
import os
import nltk
import time
import matplotlib.pyplot as plt

%matplotlib inline

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
cwd = os.getcwd() #Get current working directory
all_files = os.listdir(cwd)   # List all files in directory
strtxt = '.txt'

# To remove puctuation from the text
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
```

```python
for txtfile in all_files:
    if txtfile.__contains__(strtxt):
        file = open(txtfile, "r")
        data = file.read()
        data = data.lower()  # converting all text to lower case
        new_data = ""
        for char in data:
            if char not in punctuations:
                new_data = new_data + char
#         print(new_data+"\n")

        start = time.time()
        words = nltk.word_tokenize(new_data)
        end = time.time()
        time_withoutpreprocessing = end-start
        without_preprocessing = len(words)
        
        start = time.time()
        tokens_withpreprocessing = [lemmatizer.lemmatize(w) for w in words if not w in set(stopwords.words('english'))]
        end = time.time()
        time_withpreprocessing = end-start
        with_preprocessing = len(tokens_withpreprocessing)
        
        dataset = [without_preprocessing,with_preprocessing]
        labels = ['Without Preprocessing','With Preprocessing']
#         plt.figure(figsize=(5, 5))
#         plt.grid(color='black', linestyle='--', linewidth=2, axis='y', alpha=0.7)
        plt.bar(labels,dataset)
        # plt.xlabel("Processing Types")
        plt.ylabel("Number of tokens")
        plt.title("Processing of "+txtfile)
        plt.show()
        
        dataset = [time_withoutpreprocessing,time_withpreprocessing]
        labels = ['Without Preprocessing','With Preprocessing']
#         plt.figure(figsize=(5, 5))
        plt.bar(labels,dataset)
        # plt.xlabel("Processing Types")
        plt.ylabel("Time Value (MS)")
        plt.title("Overall time of  "+txtfile)
        plt.show()
```
